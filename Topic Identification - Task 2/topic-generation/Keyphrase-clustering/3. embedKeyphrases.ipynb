```python
import os

def read_keyphrases_from_files(folder_path):
    keyphrases_dict = {}
    for file in os.listdir(folder_path):
        art_id = os.path.splitext(file)[0]  # Assuming art_id is the filename without extension
        file_path = os.path.join(folder_path, file)
        with open(file_path, 'r', encoding='utf-8') as f:
            keyphrases = [line.strip() for line in f.readlines()]
        keyphrases_dict[art_id] = keyphrases
    return keyphrases_dict

folder_path = '../keyphrases_txt'
keyphrases_by_art_id = read_keyphrases_from_files(folder_path)
```

```python
import torch

from sentence_transformers import SentenceTransformer

cache_dir='model_cache'
model = SentenceTransformer("../mpnetv2-finetuned")

def encode(text):
    embeddings = model.encode(text)
    return embeddings
```

```python
import pickle

for art_id, keyphrases in keyphrases_by_art_id.items():
    embeddings_dict = {phrase: encode(phrase) for phrase in keyphrases}
    # Save each dictionary in a pickle file named after the art_id
    with open(f'../embeddings/embeddings_{art_id}.pkl', 'wb') as f:
        pickle.dump(embeddings_dict, f)
```

```python
import pickle
def load_embeddings(art_id):
    with open(f'../embeddings/embeddings_{art_id}.csv.pkl', 'rb') as f:
        return pickle.load(f)
```
